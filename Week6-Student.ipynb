## Cell [1]: Environment Setup & Data Loading (Slide 2)

Load the Fung-wong typhoon data and filter to the study area. Reuse your Week 5 `parse_rainfall_json()` function.

```
remove -998/zero rain stations, and convert to EPSG:3826.
Show me the top 5 stations by rain_1hr.
```

````python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import warnings
from shapely.geometry import Point
warnings.filterwarnings('ignore')

def normalize_cwa_json(data):
    """Normalize CWA JSON data to tabular format"""
    normalized = []
    for record in data['records']['Station']:
        base_info = {
            'station_id': record['StationId'],
            'station_name': record['StationName'],
            'county': record['GeoInfo']['CountyName'],
            'township': record['GeoInfo']['TownName'],
            'lat': record['GeoInfo']['Coordinates'][0]['StationLatitude'],
            'lon': record['GeoInfo']['Coordinates'][0]['StationLongitude'],
            'timestamp': record['ObsTime']['DateTime'],
            'rain_1hr': record['RainfallElement']['Past1hr']['Precipitation']
        }
        normalized.append(base_info)
    
    return pd.DataFrame(normalized)

def parse_rainfall_json(json_path):
    """Parse rainfall JSON file and return GeoDataFrame"""
    with open(json_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    df = normalize_cwa_json(data)
    
    # Create geometry points
    geometry = [Point(lon, lat) for lat, lon in zip(df['lat'], df['lon'])]
    gdf = gpd.GeoDataFrame(df, geometry=geometry, crs='EPSG:4326')
    
    return gdf

# Load and process data
print("Loading Fung-wong typhoon data...")
rain_gdf = parse_rainfall_json('data/scenarios/fungwong_202511.json')

# Filter to study area (花蓮縣 + 宜蘭縣)
study_rain = rain_gdf[rain_gdf['county'].isin(['花蓮縣', '宜蘭縣'])].copy()

# Remove stations with invalid rainfall data
study_rain = study_rain[study_rain['rain_1hr'] > 0].copy()

# Convert to EPSG:3826 (Taiwan datum)
study_rain_3826 = study_rain.to_crs('EPSG:3826')

# Extract coordinate arrays for Kriging / ML
x = study_rain_3826.geometry.x.values  # Easting (meters)
y = study_rain_3826.geometry.y.values  # Northing (meters)
z = study_rain_3826['rain_1hr'].values

print(f"Study area stations (rain > 0): {len(study_rain_3826)}")
print(f"CRS: {study_rain_3826.crs}")
print(f"\nTop 5 stations:")
print(study_rain_3826.nlargest(5, 'rain_1hr')[['station_name', 'county', 'rain_1hr']].to_string(index=False))
````

Try running Kriging directly on the raw rainfall data. **Don't worry if it looks bad — that's the point.**

**Key Concepts** (will make sense after you see the result):
- **Nugget**: How noisy is each measurement?
- **Sill**: What's the maximum difference between stations?
- **Range**: How far apart do stations need to be before they stop "knowing about each other"?

## Cell [2d]: Which Variogram Fits Best? — Range & Model Comparison (Slide 7)

Compare variograms **one variable at a time** — this is how scientists isolate effects:

- **Figure 1**: Fix model = **Spherical**, vary Range (50km, 25km, 15km) → what does Range do?

````python
# YOUR CODE HERE:
# 1. Define ranges: [50000, 25000, 15000]
# 2. Figure 1: Spherical × 3 Ranges (1×3 subplot)
#    For each: create OrdinaryKriging, plot lags vs semivariance (red dots),
#    plot fitted curve (black line), compute SSE
# 3. Figure 2: Exponential × 3 Ranges (1×3 subplot)
#    Same as above but with variogram_model='exponential'
# 4. Print summary table:
#    - Compare within Spherical (Range effect)
#    - Compare within Exponential (Range effect)
#    - Compare Spherical vs Exponential at same Range (Model effect)

ranges_m = [50000, 25000, 15000]  # meters
ranges_km = [50, 25, 15]  # for display

def spherical_model(h, sill, range_, nugget):
    """Spherical variogram model function"""
    h = np.array(h)
    result = np.where(h == 0, 0, 
                      np.where(h <= range_, 
                              nugget + sill * (1.5 * h/range_ - 0.5 * (h/range_)**3),
                              nugget + sill))
    return result

def exponential_model(h, sill, range_, nugget):
    """Exponential variogram model function"""
    h = np.array(h)
    return nugget + sill * (1 - np.exp(-3 * h / range_))

def compute_sse(ok_model, model_func):
    """Compute Sum of Squared Errors between empirical and fitted variogram"""
    empirical = ok_model.semivariance
    lags = ok_test.lags
    # Generate fitted values at the same lags
    fitted = model_func(lags, *ok_model.variogram_model_parameters)
    return np.sum((empirical - fitted) ** 2)

# ─── Figure 1: Spherical ───
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
spherical_results = []

for ax, r_m, r_km in zip(axes, ranges_m, ranges_km):
    ok_test = OrdinaryKriging(x, y, z_log, variogram_model='spherical',
                              verbose=False, enable_plotting=False, nlags=15,
                              variogram_parameters={'sill': float(z_log.var()),
                                                    'range': r_m,
                                                    'nugget': float(z_log.var() * 0.1)})
    
    # Plot empirical variogram
    ax.scatter(ok_test.lags/1000, ok_test.semivariance, c='red', s=20, alpha=0.7, zorder=5)
    
    # Plot fitted curve
    lags_fine = np.linspace(0, max(ok_test.lags), 100)
    fitted_curve = spherical_model(lags_fine, *ok_test.variogram_model_parameters)
    ax.plot(lags_fine/1000, fitted_curve, 'k-', linewidth=2)
    
    sse = compute_sse(ok_test, spherical_model)
    spherical_results.append({'Range_km': r_km, 'SSE': sse})
    
    ax.set_title(f'Spherical, Range = {r_km} km\nSSE = {sse:.3f}')
    ax.set_xlabel('Distance (km)')
    ax.set_ylabel('Semivariance')
    ax.grid(True, alpha=0.3)

plt.suptitle('Figure 1: Spherical Model - Range Comparison', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

# ─── Figure 2: Exponential ───
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
exponential_results = []

for ax, r_m, r_km in zip(axes, ranges_m, ranges_km):
    ok_test = OrdinaryKriging(x, y, z_log, variogram_model='exponential',
                              verbose=False, enable_plotting=False, nlags=15,
                              variogram_parameters={'sill': float(z_log.var()),
                                                    'range': r_m,
                                                    'nugget': float(z_log.var() * 0.1)})
    
    # Plot empirical variogram
    ax.scatter(ok_test.lags/1000, ok_test.semivariance, c='red', s=20, alpha=0.7, zorder=5)
    
    # Plot fitted curve
    lags_fine = np.linspace(0, max(ok_test.lags), 100)
    fitted_curve = exponential_model(lags_fine, *ok_test.variogram_model_parameters)
    ax.plot(lags_fine/1000, fitted_curve, 'k-', linewidth=2)
    
    sse = compute_sse(ok_test, exponential_model)
    exponential_results.append({'Range_km': r_km, 'SSE': sse})
    
    ax.set_title(f'Exponential, Range = {r_km} km\nSSE = {sse:.3f}')
    ax.set_xlabel('Distance (km)')
    ax.set_ylabel('Semivariance')
    ax.grid(True, alpha=0.3)

plt.suptitle('Figure 2: Exponential Model - Range Comparison', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

# ─── Summary Table ───
print("=== VARIOGRAM FIT COMPARISON ===")
print("\n1. WITHIN Spherical (Range effect):")
for result in spherical_results:
    print(f"   Range {result['Range_km']:2d} km: SSE = {result['SSE']:.3f}")

print("\n2. WITHIN Exponential (Range effect):")
for result in exponential_results:
    print(f"   Range {result['Range_km']:2d} km: SSE = {result['SSE']:.3f}")

print("\n3. ACROSS Models (same Range):")
for i, r_km in enumerate(ranges_km):
    sse_sph = spherical_results[i]['SSE']
    sse_exp = exponential_results[i]['SSE']
    better = "Spherical" if sse_sph < sse_exp else "Exponential"
    print(f"   Range {r_km:2d} km: Spherical={sse_sph:.3f}, Exponential={sse_exp:.3f} → {better} better")

print("\n💡 Questions:")
print("   1. Within Spherical, which Range gives the best fit?")
print("   2. Within Exponential, which Range gives the best fit?")
print("   3. At the same Range, does model choice matter much?")
````

### Cell [6]: Nearest Neighbor + IDW

Compute two additional interpolation methods so we can compare all four.

**AI Prompt Suggestion**:
```
Use scipy.interpolate.NearestNDInterpolator for nearest neighbor interpolation,
and manual IDW with scipy.spatial.distance.cdist (power=2) for IDW.
Both should predict on the same meshgrid as Kriging and RF.
Note: Don't use Rbf(function='inverse') — it's not real IDW and produces extreme values.
```

````python
from scipy.interpolate import NearestNDInterpolator
from scipy.spatial.distance import cdist

# YOUR CODE HERE:
# 1. Nearest Neighbor interpolation
# 2. IDW interpolation (手動實作，power=2)
#    ⚠️ 注意：不要用 Rbf(function='inverse')，它不是真正的 IDW，會產生極端值

nn_interp = NearestNDInterpolator(list(zip(x, y)), z)
z_nn = nn_interp(grid_xx, grid_yy)

pts = np.column_stack([x, y])
grid_pts = np.column_stack([grid_xx.ravel(), grid_yy.ravel()])
dists = cdist(grid_pts, pts)
dists[dists < 1] = 1  # 避免除以零
power = 2
weights = 1.0 / (dists ** power)
z_idw = ((weights @ z) / weights.sum(axis=1)).reshape(grid_xx.shape)

print("✓ Nearest Neighbor + IDW computed")
````

### Cell [9b]: Nugget Effect — Why Extreme Rain Gets Diluted (Slide 7)

Suao recorded **130.5 mm/hr**, but default Kriging predicts only ~71 mm at 500m away. Why?

**Nugget** controls how much the model trusts the actual measurements:
- High Nugget = "measurements are noisy" → smooths everything → extreme values diluted
- Low Nugget = "measurements are accurate" → passes through stations → extremes preserved

**Task**: Compare Nugget=10% vs Nugget=1% on a zoomed-in map around Suao. Which preserves the extreme rainfall better?

**AI Prompt Suggestion**:
```
Create two OrdinaryKriging models on z_log with identical parameters except
nugget: one with nugget = sill * 0.10, one with nugget = sill * 0.01.
Predict on a 20km×20km grid centered on the station with maximum rainfall.
Show side-by-side maps and print predicted values at 0m, 500m, 1000m, 2000m
from that station. Which Nugget preserves the extreme value better?
```

````python
# YOUR CODE HERE:
# 1. Find the station with maximum rainfall (suao_idx = np.argmax(z))
# 2. Create two OrdinaryKriging models: nugget = sill*0.10 and sill*0.01
# 3. Predict on a local grid (20km box) around that station
# 4. Plot side-by-side comparison maps
# 5. Predict at specific offsets: 0m, 500m, 1000m, 2000m from the station
# 6. Print comparison table

suao_idx = np.argmax(z)
suao_x, suao_y, suao_z = x[suao_idx], y[suao_idx], z[suao_idx]
print(f"Station with max rainfall: {suao_z:.1f} mm/hr")

sill_val = float(z_log.var())

# Create two Kriging models with different nugget values
OK_high_nugget = OrdinaryKriging(x, y, z_log, variogram_model='spherical',
                                 verbose=False, enable_plotting=False, nlags=15,
                                 variogram_parameters={'sill': sill_val,
                                                       'range': 50000.0,
                                                       'nugget': sill_val * 0.10})

OK_low_nugget = OrdinaryKriging(x, y, z_log, variogram_model='spherical',
                                verbose=False, enable_plotting=False, nlags=15,
                                variogram_parameters={'sill': sill_val,
                                                      'range': 50000.0,
                                                      'nugget': sill_val * 0.01})

# Create local grid (20km × 20km around the max station)
local_buffer = 20000  # 20km
local_resolution = 500  # 500m for finer detail

local_x_min = suao_x - local_buffer
local_x_max = suao_x + local_buffer
local_y_min = suao_y - local_buffer
local_y_max = suao_y + local_buffer

local_grid_x = np.arange(local_x_min, local_x_max, local_resolution)
local_grid_y = np.arange(local_y_min, local_y_max, local_resolution)

# Predict with both models
z_high_log, _ = OK_high_nugget.execute('grid', local_grid_x, local_grid_y)
z_low_log, _ = OK_low_nugget.execute('grid', local_grid_x, local_grid_y)

# Back-transform
z_high = np.expm1(z_high_log)
z_low = np.expm1(z_low_log)

# Plot comparison
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# High nugget (10%)
im1 = axes[0].imshow(z_high, extent=[local_x_min, local_x_max, local_y_min, local_y_max],
                     origin='lower', cmap='YlOrRd', vmin=0, vmax=suao_z*1.2)
axes[0].plot(suao_x, suao_y, 'r*', markersize=15, label=f'Max Station ({suao_z:.1f} mm/hr)')
axes[0].set_title(f'High Nugget (10% of sill)\nMore smoothing, extreme values diluted')
axes[0].legend()
plt.colorbar(im1, ax=axes[0], shrink=0.8, label='mm/hr')

# Low nugget (1%)
im2 = axes[1].imshow(z_low, extent=[local_x_min, local_x_max, local_y_min, local_y_max],
                     origin='lower', cmap='YlOrRd', vmin=0, vmax=suao_z*1.2)
axes[1].plot(suao_x, suao_y, 'r*', markersize=15, label=f'Max Station ({suao_z:.1f} mm/hr)')
axes[1].set_title(f'Low Nugget (1% of sill)\nLess smoothing, extreme values preserved')
axes[1].legend()
plt.colorbar(im2, ax=axes[1], shrink=0.8, label='mm/hr')

plt.suptitle(f'Nugget Effect on Extreme Rainfall Preservation\nAround station with {suao_z:.1f} mm/hr', 
             fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

# Predict at specific offsets from the station
offsets = [0, 500, 1000, 2000]  # meters
print(f"\nPredicted rainfall at different distances from max station ({suao_z:.1f} mm/hr):")
print(f"{'Distance (m)':<12} {'High Nugget':<12} {'Low Nugget':<12} {'Difference':<12}")
print("-" * 50)

for offset in offsets:
    # Points to the east of the station
    pred_x = suao_x + offset
    pred_y = suao_y
    
    # Predict with high nugget
    z_high_pred_log, _ = OK_high_nugget.execute('points', [pred_x], [pred_y])
    z_high_pred = np.expm1(z_high_pred_log)[0]
    
    # Predict with low nugget
    z_low_pred_log, _ = OK_low_nugget.execute('points', [pred_x], [pred_y])
    z_low_pred = np.expm1(z_low_pred_log)[0]
    
    diff = z_low_pred - z_high_pred
    print(f"{offset:<12} {z_high_pred:<12.1f} {z_low_pred:<12.1f} {diff:<12.1f}")

print(f"\n🔑 Which Nugget setting is better for CWA calibrated stations? Why?")
print("Low nugget (1%) is better for CWA calibrated stations because:")
print("- CWA stations are professionally calibrated with known measurement accuracy")
print("- Low nugget preserves real extreme rainfall events critical for flood warnings")
print("- High nugget assumes high measurement error, which may underestimate true risks")
````

⚠️ **Flip warning**: `z_kriging` row 0 = south (numpy convention). GeoTIFF row 0 = north. Use `np.flipud()` to fix.

**AI Prompt Suggestion**:
```
Save 2D numpy arrays as GeoTIFF using rasterio. I need:
- from_bounds transform with my grid extent (x_min, y_min, x_max, y_max)
- CRS = EPSG:3826, dtype = float32
- Apply np.flipud() before writing
- Save kriging_rainfall.tif, kriging_variance.tif, and rf_rainfall.tif
```

````python
import rasterio
from rasterio.transform import from_bounds

# YOUR CODE HERE:
# 1. Compute rasterio transform using from_bounds
# 2. Write a helper function save_geotiff(data, filename)
# 3. Save kriging_rainfall.tif, kriging_variance.tif, rf_rainfall.tif
# Remember: np.flipud() before writing!

transform = from_bounds(x_min, y_min, x_max, y_max,
                        width=z_kriging.shape[1], height=z_kriging.shape[0])

def save_geotiff(data, filename, crs='EPSG:3826'):
    data_flipped = np.flipud(data).astype(np.float32)
    with rasterio.open(filename, 'w', driver='GTiff',
        height=data_flipped.shape[0], width=data_flipped.shape[1],
        count=1, dtype='float32', crs=crs, transform=transform, nodata=-9999
    ) as dst:
        dst.write(data_flipped, 1)
    print(f"✓ Saved {filename}")

save_geotiff(z_kriging, 'kriging_rainfall.tif')
save_geotiff(ss_kriging, 'kriging_variance.tif')
save_geotiff(z_rf, 'rf_rainfall.tif')
````

### Cell [11]: Zonal Statistics — Township Decision Table

Compute per-township statistics from your Kriging and RF rasters, then compare them side-by-side.

**Required output**: A DataFrame with: 鄉鎮 | Kriging平均 | Kriging最大 | RF平均 | 平均variance | 可信度

````python
from rasterstats import zonal_stats

# YOUR CODE HERE:
# 1. Load township boundaries (TGOS shapefile)
# 2. Filter to 花蓮縣 + 宜蘭縣, convert to EPSG:3826
# 3. Run zonal_stats on kriging_rainfall.tif, kriging_variance.tif, rf_rainfall.tif
# 4. Create summary DataFrame with columns:
#    鄉鎮, 縣市, Kriging平均, Kriging最大, RF平均, 平均variance
# 5. Add 可信度 column:
#    HIGH: variance < 33rd percentile
#    MEDIUM: 33rd-66th percentile
#    LOW: > 66th percentile

# Note: If you don't have the township shapefile, skip this cell
# and describe what the expected output would be in the markdown below.

try:
    # Try to find township shapefile
    import os
    shapefile_paths = [
        'data/TOWN_MOI_1120725.shp',
        'data/TOWN_MOI.shp',
        '../data/TOWN_MOI_1120725.shp',
        '../../data/TOWN_MOI_1120725.shp'
    ]
    
    towns = None
    for path in shapefile_paths:
        if os.path.exists(path):
            towns = gpd.read_file(path)
            print(f"Found township shapefile: {path}")
            break
    
    if towns is None:
        raise FileNotFoundError("Township shapefile not found")
    
    # Filter to study area and convert CRS
    study_towns = towns[towns['COUNTYNAME'].isin(['花蓮縣', '宜蘭縣'])].copy()
    study_towns = study_towns.to_crs(epsg=3826)
    
    print(f"Found {len(study_towns)} townships in study area")
    
    # Run zonal statistics
    kriging_stats = zonal_stats(study_towns, 'kriging_rainfall.tif', 
                                stats=['mean', 'max'], geojson_out=False)
    variance_stats = zonal_stats(study_towns, 'kriging_variance.tif', 
                               stats=['mean'], geojson_out=False)
    rf_stats = zonal_stats(study_towns, 'rf_rainfall.tif', 
                          stats=['mean'], geojson_out=False)
    
    # Create summary DataFrame
    summary_data = []
    for i, town in study_towns.iterrows():
        summary_data.append({
            '鄉鎮': town['TOWNNAME'],
            '縣市': town['COUNTYNAME'],
            'Kriging平均': kriging_stats[i]['mean'],
            'Kriging最大': kriging_stats[i]['max'],
            'RF平均': rf_stats[i]['mean'],
            '平均variance': variance_stats[i]['mean']
        })
    
    summary_df = pd.DataFrame(summary_data)
    
    # Add confidence levels based on variance percentiles
    var_33 = summary_df['平均variance'].quantile(0.33)
    var_66 = summary_df['平均variance'].quantile(0.66)
    
    def get_confidence(var):
        if var < var_33:
            return 'HIGH'
        elif var < var_66:
            return 'MEDIUM'
        else:
            return 'LOW'
    
    summary_df['可信度'] = summary_df['平均variance'].apply(get_confidence)
    
    # Sort by Kriging average rainfall (descending)
    summary_df = summary_df.sort_values('Kriging平均', ascending=False)
    
    # Display results
    print("\n=== Township Rainfall Decision Table ===")
    print(summary_df.to_string(index=False, float_format='%.1f'))
    
    # Highlight high-risk townships
    high_risk = summary_df[(summary_df['Kriging平均'] > 50) & (summary_df['可信度'] == 'HIGH')]
    if len(high_risk) > 0:
        print(f"\n⚠️  HIGH RISK TOWNSHIPS (High Rainfall + High Confidence):")
        print(high_risk[['鄉鎮', '縣市', 'Kriging平均', '可信度']].to_string(index=False, float_format='%.1f'))
    
    # Export to CSV for further use
    summary_df.to_csv('township_rainfall_summary.csv', index=False, encoding='utf-8-sig')
    print(f"\n✓ Summary table saved to 'township_rainfall_summary.csv'")
    
except Exception as e:
    print(f"Township shapefile not found or error: {e}")
    print("\nExpected output structure:")
    print("鄉鎮 | 縣市 | Kriging平均 | Kriging最大 | RF平均 | 平均variance | 可信度")
    print("---|---|---|---|---|---|---")
    print("蘇澳鎮 | 宜蘭縣 | 85.2 | 130.5 | 78.9 | 0.12 | HIGH")
    print("五結鄉 | 宜蘭縣 | 42.1 | 65.3 | 45.7 | 0.08 | HIGH")
    print("...")
````
